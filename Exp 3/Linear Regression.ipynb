```python
import numpy as np
import matplotlib.pyplot as plt
X = [17,21,24,28,14,16,19,22,15,18] #dependant
Y = [140,189,210,240,130,100,135,166,130,170] #independant

X = np.array(X)
Y = np.array(Y)

 #Y=mX+C
n = len(X)
def MSE(A,B):
  return np.sum((A-B)**2)/n
m = 0
c = 0
loss=[]
for i in range(200):
  print(f"epoch: {i}")
  # gradient descent
  lr = 0.001

  predicted = m*X+c
  Error = MSE(predicted,Y)
  print(f"Error = {Error}")
  print(f"Values Predicted = {predicted}")
  loss.append(np.log(Error))
  m -= (2/n)*lr*np.sum((predicted-Y)*X)
  c -=  (2/n)*lr*np.sum(predicted-Y)
  print((m,c))
plt.plot(loss)
plt.show()
```
